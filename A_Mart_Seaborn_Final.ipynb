```python
#Note: fuel economy is the same as mpg, both are use interchangeably 
import seaborn as sns
```

```python
#Dataset
mpg = sns.load_dataset('mpg')
mpg
```

*Create a visual showing the relationship between horsepower and fuel economy.*

```python
#Create a visual showing the relationship between horsepower and fuel economy.
sns.relplot(data=mpg,x='horsepower',y='mpg',kind='line')
```

```python
#In general, more horsepower = lower fuel economy. This is because engines with more horsepower
#are larger, and take more fuel to generate power. 
```

*Create a table summarizing fuel economy by country of origin*

```python
#Create a table summarizing fuel economy by country of origin
mpg_per_country = mpg.groupby('origin').agg({'mpg':'mean', 'cylinders':'mean'})
mpg_per_country
```

```python
#This is average fuel economy and average number of cylinders for all vehicles
#for each country of origin. The USA tends to make more vehicles with a greater 
#number of cylinders.
```

*Create a visual showing how fuel economy has changed over time*

```python
sns.relplot(data=mpg,x='model_year',y='mpg',kind='line')
```

```python
sns.relplot(data=mpg,x='model_year',y='horsepower',kind='line')
```

```python
#The first visual shows how fuel economy has increased steadily. However, the second graph
#shows how horsepower decreased in the same way because of this.
```

*Create two other visuals or tables related to fuel economy.*

```python
#The following visual will demonstrate the relationship between a vehicle's weight
#and its fuel economy.
sns.relplot(data=mpg,x='weight',y='mpg',kind='scatter')
```

```python
#The visual shows that the weight of a vehicle substantially affects its fuel economy.
#The heavier the vehicle, the more power it expends to get moving, resulting in a lower mpg.
```

```python
#The next table will show which  vehicles have the greatest fuel economies overall.
makes = mpg.groupby('name').agg({'mpg':'mean', 'origin':'first'}).sort_values(by='mpg',ascending=False).head(15)
makes
```

```python
#The table was limited to the top 15 entries. The table shows that generally speaking, 
#European sports cars and Japanese cars tend to have the greatest fuel economies,
#making up 80% percent of the top 15.
```
